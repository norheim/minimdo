```python
from modeling.compute import Var, Par
from modeling.couplingapi import Subproblem
from modeling.execution import edges_from_components
from graph.matrixview import render_incidence
```

```python
x1,x2,x3 = Var('x_1'), Var('x_2'), Var('x_3')
```

```python
m = Subproblem()
m.add_equation(None, x1+x2)
m.add_equation(None, x3/x2)
edges = edges_from_components(m.components)
tree = {0:1, 1:1}, {}, {'x_1':1,'x_2':1,'x_3':1}
```

```python
render_incidence(edges, tree, rawvarname=True,hideticks=True, figsize=(1,1));
```

Next example

```python
a1,a2,a3,a4= Par('a_1'), Par('a_2'), Par('a_3'), Par('a_4')
A, mt = Var('A'), Var('m_t')
m = Subproblem()
Ps = m.Var('P_s', a1*A)
Pd = m.Var('P_d', a2*mt)
m.add_equation(None, Ps-Pd)
ms = m.Var('m_s', a3*mt)
m.add_equation(None, -mt+a4*A+ms);
edges = edges_from_components(m.components, include_par=False)
tree = m.Ftree, {}, {'A':1, 'm_t':1}
```

```python
render_incidence(edges, tree, rawvarname=True,hideticks=True, figsize=(4,4));
```
